```python
import cv2
import numpy as np

#Count True Positives (TP), False Positives (FP), False Negatives (FN) for a given class color.
def count_tp_fp_fn(gt_img, pred_img, color, class_index):
    # Initialize counts
    tp_count = fp_count = fn_count = 0

    # Check TP, FP, FN for each pixel
    for y in range(gt_img.shape[0]):
        for x in range(gt_img.shape[1]):
            gt_pixel = gt_img[y, x]
            pred_pixel = pred_img[y, x]

            '''
            # Print pixel colors for debugging
            print(f"gt_pixel: {gt_pixel}, pred_pixel: {pred_pixel}")
            '''
            # Check if the pixel color matches the given color (xx)
            if np.all(gt_pixel == color):
                # Check if the corresponding pixel in pred image matches the color
                if np.all(pred_pixel == color):
                    tp_count += 1  # True Positive
                else:
                    fn_count += 1  # False Negative
            # Check FP only if the pixel in pred image matches the color
            elif np.all(pred_pixel == color):
                fp_count += 1  # False Positive

    return tp_count, fp_count, fn_count


#Calculate Intersection over Union (IoU) given TP, FP, FN.
def calculate_iou(tp, fp, fn):
    if tp + fp + fn == 0:
        return 0
    else:
        return tp / (tp + fp + fn)

#Evaluate IoU for each class and print the results.
def evaluate_iou(gt_img_path, pred_img_path):
    # Read images using OpenCV
    gt_img = cv2.imread(gt_img_path,cv2.IMREAD_COLOR)
    pred_img = cv2.imread(pred_img_path,cv2.IMREAD_COLOR)

    # Define colors and class indices
    colors = [[0,0,255], [0,255,255], [255,0,0], [255, 0, 255]]
    class_indices = [0, 1, 2, 3]

    for color, class_index in zip(colors, class_indices):
        # Count TP, FP, FN
        tp, fp, fn = count_tp_fp_fn(gt_img, pred_img, color, class_index)
        print("TP:",tp," FP:",fp," FN:",fn)
        # Calculate IoU
        iou = calculate_iou(tp, fp, fn)
        # Print IoU for the class
        print(f"Class {class_index}, IoU = {iou}")

# Example usage
gt_img_path = '/content/96_groundtruth.png'
pred_img_path = '/content/255.png'
evaluate_iou(gt_img_path, pred_img_path)
```
